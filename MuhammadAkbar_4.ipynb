```python
# Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Data Pre-Processing

```python
# Import marketing campaign data file csv
df = pd.read_csv('marketing_campaign.csv')
df.head()
```

#### Melihat dan menangani data kosong

```python
df.info()
```

```python
df.isnull().sum()
```

- Jumlah baris dan kolom adalah 2240 dan 29
- Kolom Income memiliki 24 null

```python
# Menghapus data null
df.dropna(subset=['Income'])
```

- Menghapus kolom income yang bernilai null karena ada data yang kosong masih kurang dari 5%

```python
df['Income'].describe()
```

```python
df['Income']=df['Income'].fillna(df['Income'].mean())
```

- Nilai null pada kolom Income diisi dengan mean karena mempertahankan distribusi yang telah normal

```python
df.tail()
```

### Analisis Data

```python
df.describe()
```

```python
# Rata -rata
mean_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].mean()
print("Rata-rata:")
print(mean_values)
```

```python
# Task 2: Menghitung nilai modus kolom
mode_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].mode().iloc[0]
print("\nModus:")
print(mode_values)
```

```python
# Menghitung nilai median kolom
median_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].median()
print("\nMedian:")
print(median_values)
```

```python
# Task 4: Menghitung nilai range kolom
range_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].max() - df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].min()
print("\nRange:")
print(range_values)
```

```python
# Task 5: Menghitung nilai interquartile kolom
Q1 = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].quantile(0.25)
Q3 = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].quantile(0.75)
IQR = Q3 - Q1
print("\nInterquartile Range (IQR):")
print(IQR)
```

```python
# Task 6: Menghitung nilai variance kolom
variance_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].var()
print("\nVariance:")
print(variance_values)
```

```python
std_values = df[['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']].std()
print("Standar Deviasi:")
print(std_values)
```

```python
# Membuat KDE Plot
features = num
for i in range(0, len(features)):
    plt.subplot(2,4, i+1)
    sns.kdeplot(x=df[features[i]], color='green')
    plt.xlabel(features[i])
    plt.tight_layout()
```
